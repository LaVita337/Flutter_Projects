```python
import pandas as pd
from IPython.display import display
```

```python
df = pd.read_excel(r'C:\Users\USER\aiffelthon\data\1_30000_20231001.xls')
df.head()
```

```python
df = df[['표제어', '관련어']]
df.head()
```

```python
df = df.dropna(subset=['관련어'])
df.head()
```

```python
# '관련어' 열에서 \n을 제거하는 함수
def remove_example_strings(text):
    # 결측값인 경우 빈 문자열로 반환
    if isinstance(text, str):
        # \n 문자열을 공백으로 대체
        cleaned_text = text.replace('\n', '')
        return cleaned_text
    else:
        return ''

# '관련어' 열에 불용어 처리 적용
df['관련어'] = df['관련어'].apply(remove_example_strings)

# 결과 출력
df.head()
```

```python
# 행의 개수 세기
row_count = df.shape[0]

# 결과 출력
print("행의 개수:", row_count)
```

```python
# '유의어' 포함된 행의 개수 세기
synonym_count = df[df['관련어'].str.contains('유의어')].shape[0]

# '반대말' 포함된 행의 개수 세기
opposite_word_count = df[df['관련어'].str.contains('반대말')].shape[0]

# 결과 출력
print("유의어 포함된 행의 개수:", synonym_count)
print("반대말 포함된 행의 개수:", opposite_word_count)
```

```python
# '유의어' 포함된 행을 df의 형태로 모음
df_synonym = df[df['관련어'].str.contains('유의어')]
# '관련어' 열에서 '유의어'라는 단어 삭제
df_synonym['관련어'] = df_synonym['관련어'].str.replace('유의어', '')

# '반대말' 포함된 행을 df의 형태로 모음
df_opposite = df[df['관련어'].str.contains('반대말')]
# '관련어' 열에서 '반대말'이라는 단어 삭제
df_opposite['관련어'] = df_opposite['관련어'].str.replace('반대말', '')


# df_synonym과 df_opposite 데이터프레임을 각각 출력
display(df_synonym.head())
display(df_opposite.head())
```

```python
# '관련어' 열 이름을 '유의어'로 변경
df_synonym = df_synonym.rename(columns={'관련어': '유의어'})

# '관련어' 열 이름을 '반대말'로 변경
df_opposite = df_opposite.rename(columns={'관련어': '반대말'})
```

```python
# 중복 행 제거
df_synonym = df_synonym.drop_duplicates()

# 중복 행 제거
df_opposite = df_opposite.drop_duplicates()
```

```python
# df_synonym과 df_opposite 데이터프레임을 각각 출력
display(df_synonym.head())
display(df_opposite.head())
```
