```python
import os
import json

json_folder = '/content/drive/MyDrive/label'  # JSON 파일이 있는 폴더 경로
image_folder = '/content/drive/MyDrive/image_ko'  # 이미지 파일이 있는 폴더 경로

# JSON 파일 목록 가져오기
json_files = [f for f in os.listdir(json_folder) if f.endswith('.json')]

# 이미지 파일 목록 가져오기
image_files = [f for f in os.listdir(image_folder) if f.lower().endswith(('.jpg', '.jpeg', '.png', '.gif'))]

# 이미지 파일 이름 추출
image_file_names = [os.path.splitext(f)[0] for f in image_files]

# 이미지 파일 이름 집합으로 변환
image_file_name_set = set(image_file_names)

# 이미지 파일이 없는 JSON 파일 삭제
for json_file in json_files:
    json_file_path = os.path.join(json_folder, json_file)

    # JSON 파일 읽기
    with open(json_file_path, 'r') as f:
        data = json.load(f)

    # 이미지 파일 이름 가져오기
    image_file_name = data['images'][0]['file_name']

    # 이미지 파일 이름이 이미지 파일 목록에 없으면 JSON 파일 삭제
    if os.path.splitext(image_file_name)[0] not in image_file_name_set:
        os.remove(json_file_path)
        print(f"{json_file} 파일이 삭제되었습니다.")
```

```python

from PIL import Image

image_folder = '/content/drive/MyDrive/image_ko'  # 이미지 파일이 있는 폴더 경로
json_folder = '/content/drive/MyDrive/label'  # JSON 파일이 있는 폴더 경로

json_files = [f for f in os.listdir(json_folder) if f.endswith('.json')]

data2 = []
num = 0

# JSON 파일과 이미지 파일을 결합하여 데이터 생성
for json_file in json_files:
    json_file_path = os.path.join(json_folder, json_file)

    # JSON 파일 읽기
    with open(json_file_path, 'r') as f:
        data = json.load(f)

    # 이미지 파일 로드
    image_file_name = data['images'][0]['file_name']
    image_file_path = os.path.join(image_folder, image_file_name)
    image = Image.open(image_file_path)

    # JSON 데이터 파싱
    for item in data['annotations']:
        image_id = data['images'][0]['id']  # 이미지 ID 가져오기
        korean_caption = item['korean']  # 한국어 캡션 가져오기
        english_caption = item['english']  # 영어 캡션 가져오기

        # 데이터 추가
        data2.append({
            'image_id': image_id,
            'korean_caption': korean_caption,
            'english_caption': english_caption,
            'image': image,  # 이미지 파일을 추가
            'clip_embedding': num
        })
        num += 1

# 이제 data2 리스트에 필요한 데이터와 이미지가 저장되었습니다.

```

```python
import os
import json
import random
import numpy as np
from PIL import Image
import pickle
import torch
import clip
from torchvision.transforms import Compose, Resize, CenterCrop, ToTensor, Normalize

# CLIP 모델 로드
device = 'cuda' if torch.cuda.is_available() else 'cpu'
model, preprocess = clip.load('ViT-B/32', device=device)

# 이미지 전처리 함수
def preprocess_image(image_path):
    image = Image.open(image_path).convert('RGB')
    return preprocess(image)

# 데이터를 저장할 리스트
data = []

# JSON 파일 경로
json_folder = '/content/drive/MyDrive/label'

# JSON 파일 목록
json_files = [f for f in os.listdir(json_folder) if f.endswith('.json')]

# 데이터의 순서를 무작위로 섞기
random.shuffle(json_files)

for json_file in json_files:
    json_file_path = os.path.join(json_folder, json_file)

    # JSON 파일 읽기
    with open(json_file_path, 'r') as f:
        data_dict = json.load(f)

    # 이미지 파일 경로
    image_folder = '/content/drive/MyDrive/image_ko'
    image_file_name = data_dict['images'][0]['file_name']
    image_file_path = os.path.join(image_folder, image_file_name)

    # 이미지 전처리 및 임베딩 생성
    image_tensor = preprocess_image(image_file_path).unsqueeze(0).to(device)
    with torch.no_grad():
        image_features = model.encode_image(image_tensor)
    image_embedding = image_features.cpu().numpy()

    # JSON 데이터 파싱
    for item in data_dict['annotations']:
        korean_caption = item['korean']
        english_caption = item['english']

        # 데이터 추가
        data.append({
            'korean_caption': korean_caption,
            'english_caption': english_caption,
            'image_embedding': image_embedding
        })

# 피클 파일로 저장
with open('/content/drive/MyDrive/data.pkl', 'wb') as f:
    pickle.dump(data, f)
```
